```python
def square(length): # added a parameter so the square can be any size we want
    for i in range(4):
        forward(length)
        left(90)
```

```python
make_turtle()
square(30)
square(60)
```

```python
def polygon(n, length):
    angle = 360 / n
    for i in range(n):
        forward(length)
        left(angle)
```

```python
make_turtle()
polygon(7, 30)
```

```python
from jupyturtle import penup, pendown, forward, make_turtle

def jump(length):
    """Move forward length units without leaving a trail.

    Postcondition: Leaves the pen down.
    """
    penup()
    forward(length)
    pendown()
```

```python
make_turtle()
jump(20)

for i in range(10):
    jump(5)
    forward(10)
```
