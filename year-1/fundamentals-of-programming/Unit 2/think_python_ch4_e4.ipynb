```python
from jupyturtle import make_turtle, forward, left
```

```python
def parallelogram(length, angle):
    for i in range(2):
        forward(length)
        left(angle)
        forward(length)
        left(180 - angle)

def rectangle(length):
    parallelogram(length, 90)

def rhombus(length, angle):
    parallelogram(length, angle)
```

```python
make_turtle()
rectangle(60)

make_turtle()
rhombus(50, 60)
```
